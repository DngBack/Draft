```python
from spire.xls import *
from spire.xls.common import *

# Create a Workbook object
workbook = Workbook()
file = "data\Test_Big_File.xlsx"
# file = "/home/hoang.minh.tuanb@sun-asterisk.com/Downloads/0825【日経R回答】01_別紙１_①【顧】受託業務に関するチェックリスト.xlsx"
workbook.CustomFontFileDirectory = "/usr/share/fonts/truetype"
# Load an Excel document
workbook.LoadFromFile(file)

# Iterate through the worksheets in the workbook
for sheet in workbook.Worksheets: 

    # Get the PageSetup object
    pageSetup = sheet.PageSetup

    # Set page margins
    pageSetup.TopMargin = 0.3
    pageSetup.BottomMargin = 0.3
    pageSetup.LeftMargin = 0.3
    pageSetup.RightMargin = 0.3

# Set worksheet to fit to page when converting
workbook.ConverterSetting.SheetFitToPage = True
workbook.ConverterSetting.IsCellAutoFit = True

# Convert to PDF file
workbook.SaveToFile("newpdf.pdf", FileFormat.PDF)
# workbook.SaveToFile("AutoFitRowAndColumn.xlsx", ExcelVersion.Version2016)
workbook.Dispose()
```

```python
import fitz  # PyMuPDF
import pandas as pd
import os

def extract_tables_from_pdf(pdf_path, output_dir=None):
    """
    Extract tables from a PDF file and convert them to markdown.
    
    Parameters:
    - pdf_path: Path to the PDF file
    - output_dir: Directory to save markdown files (if None, will print to console)
    
    Returns:
    - A list of tuples: (page_number, table_number, markdown_text)
    """
    # Open the PDF file
    doc = fitz.open(pdf_path)
    result = []
    
    # Create output directory if it doesn't exist
    if output_dir and not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # Process each page
    for page_num, page in enumerate(doc):
        # Extract tables from the page
        tables = page.find_tables()
        # Process each table on the page
        for table_num, table in enumerate(tables):
            # Extract the table as a pandas DataFrame
            df = table.to_pandas()
            print(df)
            # Convert DataFrame to markdown
            markdown_text = df.to_markdown()
            
            # Add to results
            result.append((page_num + 1, table_num + 1, markdown_text))
            
            # Save to file if output directory is provided
            if output_dir:
                output_file = os.path.join(output_dir, 
                                          f"page_{page_num+1}_table_{table_num+1}.md")
                with open(output_file, "w", encoding="utf-8") as f:
                    f.write(markdown_text)
    
    # Close the document
    doc.close()
    return result

# Example usage
if __name__ == "__main__":
    pdf_path = "newpdf.pdf"  # Replace with your PDF file path
    output_dir = "tables_output"  # Optional: specify output directory
    
    # Extract tables and convert to markdown
    tables_markdown = extract_tables_from_pdf(pdf_path, output_dir)
    
    # # Print results
    # for page_num, table_num, markdown in tables_markdown:
    #     print(f"=== Page {page_num}, Table {table_num} ===")
    #     print(markdown)
    #     print("\n")
    
    # if output_dir:
    #     print(f"Markdown tables saved to {output_dir} directory")
```
